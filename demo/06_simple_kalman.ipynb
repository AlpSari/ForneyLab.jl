Simple Kalman filter demo
=========================

In this demo, we'll build a very simple Kalman filter using the ForneyLab toolbox. The goal is to filter noisy scalar observations.

Let `xin` be the (unknown) state of the system, `xout` the next state, `u` the system input and `y` the noisy observation of `xin`. In state-space representation, this corresponds to the very simple model `A=B=C=D=1`. For simplicity, we assume that `u` is a constant: `u_const`. We further assume that the observations are corrupted with white Gaussian noise, with variance `V_noise`. 

The Kalman filter can be represented as a Forney-style factor graph:

```
                   (u)
            (equ)   |
    (xin) --> = --> + --> (xout)
              |    (add_u_xin)
       (n)--> +(add_n_y)
              |
             (y)
```

Nodes `xin, n, y, u` hold known values. The complete Kalman filter is obtained by chaining multiple of these sections, connecting `xout` of the previous stage to `xin` of the next stage. 

Let's start by importing ForneyLab, defining the constants and creating the nodes:

````julia
using ForneyLab

# Settings
V_noise = 200.0 # variance of the noise (n)
u_const = 1.0   # fixed value for u

# Create nodes
xin         = TerminalNode(name="xin")
xout        = TerminalNode(name="xout")
n           = TerminalNode(GaussianDistribution(m=0.0, V=V_noise), name="n")
y           = TerminalNode(name="y")
add_n_y     = AdditionNode(name="add_n_y")
equ         = EqualityNode(name="equ")
add_u_xin   = AdditionNode(name="add_u_xin")
u           = TerminalNode(GaussianDistribution(m=u_const, V=0.0), name="u");
````

As you can see, we create `TerminalNode` nodes for the nodes that hold known values or distributions. We can pass the value of a `TerminalNode` to the constructor, as we do for the `n` and `u` nodes. These nodes will always send out a message with this payload. The `GaussianDistribution` is defined by its mean vector (`m`) and covariance matrix (`V`). The message of the `u` node has zero variance since we have no uncertainty about its value. Further, we have 2 addition nodes and 1 equality constraint node with 3 interfaces.

Now let's connect the nodes by defining the edges, and define a time wrap to connect the sections:

````julia
# Create edges
Edge(xin.out, equ.interfaces[1])
Edge(n.out, add_n_y.in1)
Edge(y.out, add_n_y.in2)
Edge(add_n_y.out, equ.interfaces[2])
Edge(equ.interfaces[3], add_u_xin.in2)
Edge(u.out, add_u_xin.in1)
Edge(add_u_xin.out, xout)
# Define time wrap
addTimeWrap(xout, xin); # xin[n] = xout[n-1];
````

Now that the factor graph is complete, we can set the initial message, create the message passing schedule, and create a buffer to hold the results. In addition to the write buffer, which was introduced in the previous demo, we also define a read buffer that reads an input value on each call to `step()`.

````julia
# Set initiatal value of xin to get started
xin.value = GaussianDistribution(m=1.0+randn()*sqrt(V_noise), V=V_noise)
schedule = generateSchedule!(add_u_xin.out)

# Data
n_samples = 100
noisy_y = [i+randn()*sqrt(V_noise) for i=1:n_samples] # Generate noisy observations of y
y_buffer = [GaussianDistribution(m=noisy_y[i], V=0.0) for i=1:n_samples] # Convert scalar observations to Gaussians with 0 variance
setReadBuffer(y, y_buffer) # TerminalNode y will take value y_buffer[n] in timestep n

# Create output buffer to record the values we wish to observe in every timestep
state = setWriteBuffer(add_u_xin.out)

show(schedule)
````

All set. Now we apply the filter by calculating the message towards `xout`. Then, we move this message to `xin` and repeat the process 100 times. This way, we simulate a forward pass over 100 sections.

````julia
# Do forward passes
while !isempty(y_buffer)
    step() 
end
````

That was easy!
Let's plot the filter output (mean and standard deviation of `xout`) together with the noisy observations to see if it worked.

````julia
using PyPlot
results_mean = Float64[mean(s)[1] for s in state]
results_variance = Float64[var(s)[1,1] for s in state]
plot([1:n_samples], noisy_y, "b*")
plot([1:n_samples], results_mean, "b-")
fill_between([1:n_samples], results_mean-sqrt(results_variance), results_mean+sqrt(results_variance), color="b", alpha=0.3);
grid(true)
````

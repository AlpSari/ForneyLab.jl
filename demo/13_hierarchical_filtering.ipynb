#Hierarchical filtering

This demo demonstrates a hierarchical filtering implementation with ForneyLab. We model a layered random walk, where the emission of a higher layer determines the precision of a lower layer. This demo is inspired by the work of (Mathys, 2012).

```julia
using PyPlot
using ForneyLab
```

##The generative model

We build a section of the generative model as shown below:

<img src="files/images/hierarchical_filter_generative.png">

```julia
# Layer 1
l1_x_k_min_node = PriorNode(name="l1_x_k_min") # Observed signal
l1_x_k_node = TerminalNode(vague(GaussianDistribution), name="l1_x_k")
l1_add_node = AdditionNode(name="l1_add")
l1_exp_node = ExponentialNode(name="l1_exp")
l1_g_node = GaussianNode(form="precision", m=0.0, name="l1_g")
Edge(l1_x_k_min_node.out, l1_add_node.in1)
Edge(l1_add_node.out, l1_x_k_node.out)
Edge(l1_exp_node.out, l1_g_node.precision, GammaDistribution)
Edge(l1_g_node.out, l1_add_node.in2, GaussianDistribution)

# Layer 2
l2_x_k_min_node = PriorNode(name="l2_x_k_min")
l2_x_k_node = TerminalNode(vague(GaussianDistribution), name="l2_x_k")
l2_add_node = AdditionNode(name="l2_add")
l2_eq_node = EqualityNode(name="l2_eq")
l2_exp_node = ExponentialNode(name="l2_exp")
l2_g_node = GaussianNode(form="precision", m=0.0, name="l2_g")
Edge(l2_eq_node.interfaces[3], l1_exp_node.in1)
Edge(l2_add_node.out, l2_eq_node.interfaces[1])
Edge(l2_eq_node.interfaces[2], l2_x_k_node.out)
Edge(l2_x_k_min_node.out, l2_add_node.in1)
Edge(l2_exp_node.out, l2_g_node.precision, GammaDistribution)
Edge(l2_g_node.out, l2_add_node.in2, GaussianDistribution)

# Layer 3
theta_k_min_node = PriorNode(name="theta_k_min")
theta_k_node = TerminalNode(vague(GaussianDistribution), name="theta_k")
l3_eq_node = EqualityNode(name="l3_eq")
Edge(l3_eq_node.interfaces[3], l2_exp_node.in1)
Edge(l3_eq_node.interfaces[2], theta_k_node.out)
Edge(theta_k_min_node.out, l3_eq_node.interfaces[1]);
```

##Sample generation

We assume the prior parameters fixed (grey). To draw samples from this model we use an ancestral sampling procedure. We draw a sample after each section and use the result as input for the next section.

We need to additionally assume that lower levels do not influence higher levels. We do this by placing an uninformative message at the interfaces denoted with a black dot. 

<img src="files/images/hierarchical_filter_sample.png">

```julia
n_samples = 100

# Connect gain chains
wrap(node("theta_k"), node("theta_k_min"))
wrap(node("l2_x_k"), node("l2_x_k_min"))
wrap(node("l1_x_k"), node("l1_x_k_min"))

# Set initial known parameters
theta_0 = DeltaDistribution([5.2])
l2_x_0 = DeltaDistribution([12.7])
l1_x_0 = DeltaDistribution([0.0]) # Start walk at origin
node("theta_k_min").value = theta_0
node("l2_x_k_min").value = l2_x_0
node("l1_x_k_min").value = l1_x_0

# Buffers for storing results
theta_x = setWriteBuffer(node("theta_k").out.partner)
l2_x = setWriteBuffer(node("l2_x_k").out.partner)
l1_x = setWriteBuffer(node("l1_x_k").out.partner)

# Assume higher layers have no knowledge of lower layers (and note that the reverse does not hold) 
setMessage!(node("l2_exp").in1, Message(vague(GaussianDistribution)))
setMessage!(node("l1_exp").in1, Message(vague(GaussianDistribution)))

# draw() # Uncomment to inspect graph

# Initialize algorithm
algo = SumProduct.Algorithm()
schedule = algo.fields[:schedule]

# Define where to draw samples and pass expectations
setPostProcessing!(schedule, node("theta_k").out.partner, sample)
setPostProcessing!(schedule, node("l2_x_k").out.partner, sample)
setPostProcessing!(schedule, node("l1_x_k").out.partner, sample)
setPostProcessing!(schedule, node("l2_exp").out, expectation)
setPostProcessing!(schedule, node("l1_exp").out, expectation)

# Estimate
for i = 1:n_samples
    step(algo);
end
```

With the estimation done, we can plot the result of our efforts.

```julia
# Plot the samples
x_vals = [mean(x_k)[1] for x_k in l1_x]
scatter(1:n_samples, x_vals)
xlabel("k")
ylabel("x[k]")
grid(true)
```

#Gain estimation

From the previously generated samples we can attempt to re-estimate the gain by employing variational message passing. We start again with the generative model. We assume broad priors on the initial gains and fix the previously drawn samples.

<img src="files/images/hierarchical_filter_gain.png">

```julia
# Reset the wraps and buffers
clearMessages!()
clearWraps()
clearBuffers()

# Connect gain chains
wrap(node("theta_k"), node("theta_k_min"))
wrap(node("l2_x_k"), node("l2_x_k_min"))

# Read from our samples
setReadBuffer(node("l1_x_k_min"), deepcopy([l1_x_0, l1_x[1:end-1]]))
l1_x_buffer = setReadBuffer(node("l1_x_k"), deepcopy(l1_x))

# Buffers for storing results
theta_x_est = setWriteBuffer(node("theta_k").out.partner)
l2_x_est = setWriteBuffer(node("l2_x_k").out.partner)

# Set the priors
# Note that these were previously the known parameters.
# We now define a broad prior to denote our limited knowledge.
theta_0_prior = GaussianDistribution(m=0.0, V=1000.0)
l2_x_0_prior = GaussianDistribution(m=0.0, V=1000.0)
node("theta_k_min").value = theta_0_prior
node("l2_x_k_min").value = l2_x_0_prior

# At this point we cannot assume higher layers have no knowledge of lower layers.
# We need to propagate information upwards from the samples towards layer two and theta, so we use variational message passing.
# We define a variational inference procedure using a mean field factorization.
algo = VMP.Algorithm(n_iterations = 10)

# for subgraph in algo.fields[:factorization].factors # Uncomment to inspect graph
#     draw(subgraph)
# end

run(algo)
```

We can show the parameter estimation process by plotting the state estimation after each step.

```julia
# Plot the results in an orderly fashion
# True values
theta_m = Float64[mean(theta_x_k)[1] for theta_x_k in theta_x]
l2_x_m = Float64[mean(l2_x_k)[1] for l2_x_k in l2_x]

# Our online estimation
theta_est_m = Float64[mean(theta_est_k)[1] for theta_est_k in theta_x_est]
theta_est_V = Float64[var(theta_est_k)[1,1] for theta_est_k in theta_x_est]
l2_x_est_m = Float64[mean(l2_x_est_k)[1] for l2_x_est_k in l2_x_est]
l2_x_est_V = Float64[var(l2_x_est_k)[1,1] for l2_x_est_k in l2_x_est]

# Plot
plot(1:n_samples, theta_est_m, color="green", label="theta_est")
plot(1:n_samples, theta_m, linestyle="--", color="green", label="theta")
fill_between(1:n_samples, theta_est_m-sqrt(theta_est_V), theta_est_m+sqrt(theta_est_V), color="green", alpha=0.2)
plot(1:n_samples, l2_x_est_m, color="blue", label="l2_x_est")
plot(1:n_samples, l2_x_m, linestyle="--", color="blue", label="l2_x")
fill_between(1:n_samples, l2_x_est_m-sqrt(l2_x_est_V), l2_x_est_m+sqrt(l2_x_est_V), color="blue", alpha=0.2)
xlabel("k")
ylabel("est[k]")
grid(true)
legend()
title("State estimation and true values over k");
```

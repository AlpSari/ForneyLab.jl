#Online prediction demo

In this demo we use a hierarchical Gaussian filter to learn from and make online predictions for real data. We use 180 dayly measurements of the currency exchange between euros and dollars. For each day we execute predict-measure-learn loop. More specifically, we

1. predict the exchange rate for the next day based on previous samples and our hidden state estimation;
2. observe the value for the next day and compare it with our estimation;
3. update the hidden state based on the newly observed sample.

We start again by building the generative model.

```julia
using PyPlot
using ForneyLab

# Build the generative model
graph = FactorGraph()

# Layer 1
l1_x_k_min_node = TerminalNode(name="l1_x_k_min") # Observed signal
l1_x_k_node = TerminalNode(vague(GaussianDistribution), name="l1_x_k")
l1_add_node = AdditionNode(name="l1_add")
l1_log_node = LogarithmicNode(name="l1_log")
l1_g_node = GaussianNode(form="precision", m=0.0, name="l1_g")
Edge(l1_x_k_min_node.out, l1_add_node.in1)
Edge(l1_add_node.out, l1_x_k_node.out)
Edge(l1_log_node.out, l1_g_node.precision, GammaDistribution)
Edge(l1_g_node.out, l1_add_node.in2, GaussianDistribution)

# Layer 2
l2_x_k_min_node = TerminalNode(name="l2_x_k_min")
l2_x_k_node = TerminalNode(vague(GaussianDistribution), name="l2_x_k")
l2_add_node = AdditionNode(name="l2_add")
l2_eq_node = EqualityNode(name="l2_eq")
l2_log_node = LogarithmicNode(name="l2_log")
l2_g_node = GaussianNode(form="precision", m=0.0, name="l2_g")
Edge(l2_eq_node.interfaces[3], l1_log_node.in1)
Edge(l2_add_node.out, l2_eq_node.interfaces[1])
Edge(l2_eq_node.interfaces[2], l2_x_k_node.out)
Edge(l2_x_k_min_node.out, l2_add_node.in1)
Edge(l2_log_node.out, l2_g_node.precision, GammaDistribution)
Edge(l2_g_node.out, l2_add_node.in2, GaussianDistribution)

# Layer 3
theta_k_min_node = TerminalNode(name="theta_k_min")
theta_k_node = TerminalNode(vague(GaussianDistribution), name="theta_k")
l3_eq_node = EqualityNode(name="l3_eq")
Edge(l3_eq_node.interfaces[3], l2_log_node.in1)
Edge(l3_eq_node.interfaces[2], theta_k_node.out)
Edge(theta_k_min_node.out, l3_eq_node.interfaces[1]);
```

Next we load and visualize the dataset. These are the dayly exchance rates from dollars to euros from June 22 to December 18, 2014, obtained from `www.oanda.com`.

```julia
# Load the dataset
data = readdlm("data/dollar_euro_exchange_rate.csv", ';')
x = reverse(data[:,2]) # Extract values and order from past to present
N = length(x)

# Visualize the data
scatter(1:N, x)
xlabel("day")
ylabel("exchange rate")
grid(true)
title("Exchange rate from dollars to euros");
```

#Predict

We assume broad priors and prepare the graph for a first prediction.

```julia
# State priors
theta_0 = GaussianDistribution(m=0.0, V=1000.0)
l2_x_0 = GaussianDistribution(m=0.0, V=1000.0)
theta_k_min_node.value = theta_0
l2_x_k_min_node.value = l2_x_0

# Initialize read buffer
x_k_min_buffer = setReadBuffer(l1_x_k_min_node, deepcopy(x))
x_k_est_buffer = setWriteBuffer(l1_x_k_node.out.partner)

# Generate a schedule
generateSchedule!(l1_x_k_node.out.partner)
predict_schedule = graph.factorization[1].internal_schedule

# Get expectations where required
setSummaryOperation!(predict_schedule, findNodeByName("l2_log").out, :sumproduct_expectation)
setSummaryOperation!(predict_schedule, findNodeByName("l1_log").out, :sumproduct_expectation)

# Predict without learning
while !isempty(x_k_min_buffer)
    step()
end
```

```julia
# Plot naive prediction results
x_est_m = Float64[mean(x_k_est)[1] for x_k_est in x_k_est_buffer]
x_est_V = Float64[var(x_k_est)[1,1] for x_k_est in x_k_est_buffer]
plot(2:N+1, x_est_m, color="red", label="estimated rate")
fill_between(1:N, x_est_m-sqrt(x_est_V), x_est_m+sqrt(x_est_V), color="red", alpha=0.2)
plot(1:N, x, linestyle="--", color="blue", label="actual rate")
xlabel("day")
ylabel("exchance rate")
grid(true)
legend()
title("Exchange rate prediction and actual value for consecutive days without learning");
```

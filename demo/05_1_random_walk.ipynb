Online state estimation for random walk example
===
This demo implements the state estimation for a random walk with Gaussian noise. Along the way we will introduce the `TimeWrap` concept, `step()` function and the use of output buffers. 

Let our estimation of the state `X(n)` be a univariate Gaussian distribution with mean `m` and variance `V` at time `n`. At each consecutive timestep we update `X(n)` by adding a constant Gaussian noise `N`, with mean `m_n` and variance `V_n`. We want to estimate the mean and variance for `X` at a final timestep `n=n_steps`. 

```
           [N]
            |   
    X(n-1)  v  X(n)
   ------->[+]----->
```

We start by including the ForneyLab.jl toolbox and initializing these basic settings.

````julia
# Load ForneyLab.jl
using ForneyLab

# Basic settings
n_steps = 50  # Total number of timesteps.
m_noise = 0.0 # Mean of the noise, corresponds with a drift of X.
V_noise = 1.0 # Variance of the Gaussian noise.
m_0     = 0.0 # Initial setting for the state X at n=0;
````

The graph above shows one step in the random walk process. There are two ways to set up the nodes. Either initialize one timestep and feed the output back to the input, or build a chain by initializing all timesteps and connecting them to form the complete graph. We will choose the first option in this demo.

````julia
# Initialize and connect nodes
X_n_min_1_node = TerminalNode(GaussianDistribution(m=m_0, V=0.0), name="X_n_min_1") # Initial state
X_n_node       = TerminalNode(name="X_n")
N_node     = TerminalNode(GaussianDistribution(m=m_noise, V=V_noise), name="N") # Constant noise
add_node       = AdditionNode()

Edge(X_n_min_1_node.out, add_node.in1)
Edge(N_node.out, add_node.in2)
Edge(add_node.out, X_n_node.out);
````

Now we need to indicate that the output at the right is actually the input for the next section. We do so by setting a `TimeWrap`:

````julia
addTimeWrap(X_n_node, X_n_min_1_node);
````

Now when we visualize the graph, we notice the green arrow indicating the presence of a time wrap.

````julia
graphViz()
````

The quantity we are interested in is the state of `x`. We can tell ForneyLab to store the online state estimate for each timepoint by initializing a write buffer. We also need to generate a schedule. With the write buffer set, `generateSchedule()` will automatically generate a schedule to calculate the message towards the interface of this buffer.

````julia
X = setWriteBuffer(add_node.out)
generateSchedule!();
````

Now our graph is ready for state estimation. We can invoke the `step()` function that will execute the schedule, write to the buffer and reset the graph for the next time step.

````julia
for n = 1:n_steps
    step()
end
````

Our calculated state has been written to our buffer. We can extract the mean and standard deviation and plot the result.

````julia
# Extract mean and standard deviation
X_means = [mean(X_n)[1] for X_n in X]
X_sds = [sqrt(var(X_n)[1,1]) for X_n in X]

# Make plot
using PyPlot
plot([1:n_steps], X_means, "b-")
fill_between([1:n_steps], X_means-X_sds, X_means+X_sds, color="b", alpha=0.3)
xlabel("n")
ylabel("X[n]")
grid(true)
````

Simple Kalman filter demo
=========================

In this demo, we'll build a very simple Kalman filter using the ForneyLab toolbox. The goal is to filter noisy scalar observations.

Let `xin` be the (unknown) state of the system, `xout` the next state, `u` the system input and `y` the noisy observation of `xin`. In state-space representation, this corresponds to the very simple model `A=B=C=D=1`. For simplicity, we assume that `u` is a constant: `u_const`. We further assume that the observations are corrupted with white Gaussian noise, with variance `V_noise`. 

The Kalman filter can be represented as a Forney-style factor graph:

```
                   (u)
            (equ)   |
    (xin) --> = --> + --> (xout)
              |    (add_u_xin)
       (n)--> +(add_n_y)
              |
             (y)
```

Nodes `xin, n, y, u` hold known values. The complete Kalman filter is obtained by chaining multiple of these sections, connecting `xout` of the previous stage to `xin` of the next stage. 

Let's start by importing ForneyLab, defining the constants and creating the nodes:

````julia
using ForneyLab

# Settings
V_noise = [200.0] # variance of the noise (n)
u_const = [1.0]   # fixed value for u

# Create nodes
xin         = ConstantNode(name="xin")
n           = ConstantNode(GaussianMessage(m=[0.0], V=V_noise), name="n")
y           = ConstantNode(name="y")
add_n_y     = AdditionNode(name="add_n_y")
equ         = EqualityNode(3, name="equ") # equ node with 3 interfaces
add_u_xin   = AdditionNode(name="add_u_xin")
u           = ConstantNode(GaussianMessage(m=u_const, V=[0.0]), name="u");
````

As you can see, we create `ConstantNode` nodes for the nodes that hold known values or distributions. We can pass the value of a `ConstantNode` to the constructor, as we do for the `n` and `u` nodes. These nodes will always send out this message. The `GaussianMessage` is defined by its mean vector (`m`) and covariance matrix (`V`). Alternatively, one can also define a weighted mean vector (`xi`) or a precision matrix (`W`). The message of the `u` node has zero variance since we have no uncertainty about its value. Further, we have 2 addition nodes and 1 equality constraint node with 3 interfaces, sometimes also called ports. As can be seen, we can (optionally) define names for nodes, which enables easy debugging.

Now let's connect the nodes by defining the edges:

````julia
# Create edges
Edge(xin.out, equ.interfaces[1])
Edge(n.out, add_n_y.in1)
Edge(y.out, add_n_y.in2)
Edge(add_n_y.out, equ.interfaces[2])
Edge(equ.interfaces[3], add_u_xin.in2)
Edge(u.out, add_u_xin.in1);
````

An edge is directed, from tail interface (first argument) to head. For most node types, the interfaces have easy names. For example, a `ConstantNode` has just one interface: `xin.interface`. An addition node has 3: `in1`, `in2`, and `out`. The interfaces of a node are documented in the header of the file that defines the node.

Now that the factor graph is complete, we can set the initial message and create variables to hold the results:

````julia
# Set initiatal value of xin to get started
xin.value = GaussianMessage(m=[1.0+randn()*sqrt(V_noise)], V=V_noise)

# Arrays to hold the results (mean and variance of xout after each section)
noisy_y = Array(Float64, 100)
results_mean = Array(Float64, 100)
results_variance = Array(Float64, 100);
````

All set, now we apply the filter by calculating the message towards `xout`. Then, we move this message to `xin` and repeat the process 100 times. This way, we simulate a forward pass over 100 sections.

````julia
# Do forward passes
for i=1:100
    # Setup noisy observation y
    noisy_y[i] = (i+randn()*sqrt(V_noise))[1]
    y.value = GaussianMessage(m=[noisy_y[i]], V=[0.0])
    # Calculate the outbound message on add_u_xin.out
    # This will recurse through the graph to calculate all required messages
    msg = calculateMessage!(add_u_xin.out)
    # Make sure we have a (mean, variance) parametrization of the message
    # This will convert msg to (mean, variance) parametrization if required
    ensureMVParametrization!(msg)
    # Save the result
    results_mean[i] = msg.m[1]
    results_variance[i] = msg.V[1,1]
    # Move xout to xin for the next section
    xin.value = deepcopy(msg)
end
````

That was easy!
Let's plot the filter output (mean and standard deviation of `xout`) together with the noisy observations to see if it worked.

````julia
using Gadfly
set_default_plot_size(9inch, 9inch/golden)
plot(layer(x=[1:100], ymin=results_mean-sqrt(results_variance), ymax=results_mean+sqrt(results_variance), Geom.ribbon), layer(x=[1:100], y=noisy_y, Geom.point), layer(x=[1:100], y=results_mean, Geom.line), Guide.xlabel("sample"))
````

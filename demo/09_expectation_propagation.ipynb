```julia
using ForneyLab
```

```julia
# Build graph
g = FactorGraph()
t_real = TerminalNode(GaussianDistribution(m=0.0, V=20.0))
sigmoid = SigmoidNode()
t_bool = TerminalNode(DeltaDistribution(false))
e_real = Edge(t_real, sigmoid.i[:real], GaussianDistribution)
Edge(sigmoid.i[:bool], t_bool, BernoulliDistribution)
draw()
```

```julia
schedule = convert(Schedule, [t_real.i[:out], t_bool.i[:out]], sumProduct!)
push!(schedule, ScheduleEntry(sigmoid.i[:real], expectation!))
show(schedule)
```

```julia
run(Algorithm(schedule))
show(ensureMVParametrization!(e_real.marginal))
```

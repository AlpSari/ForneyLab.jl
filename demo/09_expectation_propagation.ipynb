```julia
using ForneyLab
```

```julia
# Build graph
g = FactorGraph()
t_real = TerminalNode(GaussianDistribution(m=0.0, V=20.0))
sigmoid = SigmoidNode()
t_bool = TerminalNode(DeltaDistribution(false))
e_real = Edge(t_real, sigmoid.i[:real], GaussianDistribution)
Edge(sigmoid.i[:bin], t_bool, BernoulliDistribution)
draw()
```

```julia
algo = ExpectationPropagation.Algorithm([sigmoid.i[:real]], num_iterations=3)
println("Pre-sites schedule:")
show(algo.fields[:pre_sites_schedule])
println("\nCavities schedule:")
show(algo.fields[:cavities_schedule])
println("\nSites schedule:")
show(algo.fields[:sites_schedule])
# schedule = convert(Schedule, [t_real.i[:out], t_bool.i[:out]], sumProduct!)
# push!(schedule, ScheduleEntry(sigmoid.i[:real], ep!))
# show(schedule)
```

```julia
run(algo)
calculateMarginal!(e_real)
show(ensureMVParametrization!(e_real.marginal))
```

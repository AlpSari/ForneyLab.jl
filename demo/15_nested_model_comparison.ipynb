# Nested model comparison with the Savage-Dickey ratio

The Savage-Dickey ratio for model comparison (Dickey, 1971; Penny and Ridgway, 2013) is used to as a way to calculate the Bayes factor for nested models. In this demo we first derive the Savage-Dickey ratio, after which we illustrate its usage in ForneyLab.

The quality of a simple model $M$ with respect to measured data $D$ is given by Bayes rule:

\begin{equation*}
    p(M|D) = \frac{p(D|M)p(M)}{p(D)}\,.
\end{equation*}

It is hard to calculate this probability directly, because evaluation of $p(D)$ required integration over all possible models. In practice we want to compare the performance of a simple model $M_s$ with a more complex model $M_c$, in which case it suffices to evaluate their relative performance on the data:

\begin{equation}
    \frac{p(M_s|D)}{p(M_c|D)} = \frac{ \frac{p(D|M_s)p(M_s)}{p(D)} }{ \frac{p(D|M_c)p(M_c)}{p(D)} } = \frac{p(D|M_s)}{p(D|M_c)} \times\frac{p(M_s)}{p(M_c)} = BF_{sc} \times \frac{p(M_s)}{p(M_c)}\,,
\end{equation}

where the first term $BF_{sc}$ is defined as the Bayes factor. The Bayes factor expresses the likelihood ratio of two models, indicating which model best fits the data. The second term is the prior belief ratio over our models, which is set to $1$ when we have no prior preference (a preference independent of the data) of one model over the other.

## Expressing the Bayes factor as the Savage-Dickey ratio

When the simpler model $M_s$ is nested in the more complex model $M_c$, we can write the Bayes factor as the Savage-Dickey ratio, which we derive below (Penny and Ridgway, 2013). Nested here means that the parameters $\{\theta_s\}$ that describe $M_s$ are a subset of the parameters $\{\theta_s, \theta_c\}$ that decribe $M_c$. Therefore the likelihoods for these models compare as

\begin{equation}
    p(D|\theta_s, M_s) = p(D|\theta_s, \theta_c=0, M_c)\,.
\end{equation}

It is important to understand this reasoning, because it allows us to write the $p(D|M_1)$ term in the Bayes factor as:

\begin{align*}
    p(D|M_s) &= \int p(D|\theta_s, M_s) \times p(\theta_s|M_s) \operatorname{d}\!\theta_s\\
    &= \int p(D|\theta_s, \theta_c=0, M_s) \times p(\theta_s|M_s) \operatorname{d}\!\theta_s\\
    &= \int p(D|\theta_s, \theta_c=0, M_c) \times p(\theta_s|M_c) \operatorname{d}\!\theta_s\,,
\end{align*}

where we set the prior over $\theta_s$ the same in both models: $p(\theta_s|M_s) = p(\theta_s|M_c)$. We integrate and apply Bayes rule a second time:

\begin{align*}
    p(D|M_s) &= p(D|\theta_c=0, M_c)\\
    &= \frac{p(\theta_c=0|D, M_c)p(D|M_c)}{p(\theta_c=0|M_c)}\,.
\end{align*}

When we substitute this expression back into the Bayes factor, we see something interesting:

\begin{align*}
    \frac{p(D|M_s)}{p(D|M_c)} = \frac{p(\theta_c=0|D, M_c)}{p(\theta_c=0|M_c)}\,,
\end{align*}

which is known as the Savage-Dickey ratio.

## Using the Savage-Dickey ratio in practice

The Savage-Dickey ratio is very useful in practice. It says that when we have a nested model where we set the same priors in both models for the overlapping parameters, then we can calculate the Bayes factor by assessing how much the data tells us about the necessity of the additional parameters. For example, if the provided data give us reason to believe that the additional parameters in the more complicated model are required to explain the data better:

\begin{equation}
    p(\theta_c=0|D, M_c) < p(\theta_c=0|M_c)\,,
\end{equation}

then we should favour the more complicated model. Because of numerical stability and intuitive reasons it is more practical to compare the logarithms of the probabilities and communicate the difference in evidence for the two models in terms of decibels:

\begin{equation}
    10 \operatorname{log}_{10} BF_{sc} = 10 \operatorname{log}_{10} p(\theta_c=0|D, M_c) - 10 \operatorname{log}_{10} p(\theta_c=0|M_c)
\end{equation}

We will now examine a practical case where we estimate the parameters for a Gaussian random walk model with step precision $\gamma$ and drift $m$. We compare a model $M_d$ that incorporates the drift with a simpler nested model $M_s$ that ignores the drift. The factor graphs for both models are shown below:

<img src="files/images/15_random_walk_models.png">

We start by including ForneyLab and generating our training data.

```julia
using PyPlot
using ForneyLab

gamma = 10.0 # Step precision
gam_0 = GammaDistribution(a=1.0, b=1.0/9.0) # Prior for step precision used in parameter estimation
m = 0.2 # Drift

N = 50 # Number of steps
x_vals = [0.0] # Starting position

for step=1:N # Do the random walk
    push!(x_vals, x_vals[end] + sqrt(1.0/gamma)*randn() + m)
end

# Incorporate measurement errors in data
x = [GaussianDistribution(m=x_k_val, V=tiny()) for x_k_val in x_vals]

scatter(0:N, x_vals) # Plot the data
grid()
```

## Parameter estimation

In order to calculate the Savage-Dickey ratio we only need to build the more complex model $M_d$. We write a function to construct the factor graph and prepare the model for parameter estimation (illustrated below):

<img src="files/images/15_random_walk_param_est.png">

```julia
function estimateDrift(x::Array{GaussianDistribution, 1}, gam_0::GammaDistribution)
    # Estimates the drift from the more complicated model
    
    FactorGraph() # Start a new factor graph

    # Nodes
    g_node = GaussianNode(form="precision")
    add_node = AdditionNode()
    x_k_min_node = TerminalNode()
    x_k_node = TerminalNode()
    m_eq_node = EqualityNode()
    m_min_node = TerminalNode()
    m_node = TerminalNode(vague(GaussianDistribution))
    gam_eq_node = EqualityNode()
    gam_min_node = TerminalNode()
    gam_node = TerminalNode(vague(GammaDistribution))

    # Edges
    Edge(m_eq_node.interfaces[3], g_node.mean, GaussianDistribution)
    Edge(gam_eq_node.interfaces[3], g_node.precision, GammaDistribution)
    Edge(g_node.out, add_node.in2, GaussianDistribution)
    Edge(x_k_min_node.out, add_node.in1)
    Edge(add_node.out, x_k_node.out)
    Edge(m_min_node.out, m_eq_node.interfaces[1])
    Edge(gam_min_node.out, gam_eq_node.interfaces[1])
    Edge(m_eq_node.interfaces[2], m_node.out)
    Edge(gam_eq_node.interfaces[2], gam_node.out)

    # Chain the parameters
    addTimeWrap(m_node, m_min_node)
    addTimeWrap(gam_node, gam_min_node)

    # Define a mean field factorization
    factorizeMeanField!()
    setVagueMarginals!()

    # Set the priors
    m_min_node.value = vague(GaussianDistribution) # Initialize additional parameter as unknown
    gam_min_node.value = gam_0 # Gamma distribution prior value

    # Initialize buffers
    x_buffer = setReadBuffer(x_k_min_node, deepcopy(x[1:end-1]))
    setReadBuffer(x_k_node, deepcopy(x[2:end]))
    m_est = setWriteBuffer(m_node.out.partner)
    gam_est = setWriteBuffer(gam_node.out.partner)

    # Initialize schedule
    generateSchedule!()

    # Estimate
    while !isempty(x_buffer)
        step(n_iterations = 50)
    end

    # Return the estimated value for the drift
    return m_est[end]
end;
```

We can now estimate the value for the drift from our data and constructed model $p(m|x, M_d)$ by simply calling `estimateDrift()` on the provided data:

```julia
using Distributions

# Infer the drift
p_m_given_x_M_d = estimateDrift(x, gam_0)
println("Estimate for p(m | x, M_d):\n", p_m_given_x_M_d) # Print estimation result
```

But in order to evaluate relative model performance:

\begin{equation}
    10 \operatorname{log}_{10} BF_{sd} = 10 \operatorname{log}_{10} p(m=0|x, M_d) - 10 \operatorname{log}_{10} p(m=0|M_d)\,,
\end{equation}

we need to calculate $p(m|M_d)$ as well. This can be easily achieved by feeding uninformative data to the model. 

```julia
# Construct an uninformative data set
x_uninf = [vague(GaussianDistribution)]
for step=1:N
    push!(x_uninf, vague(GaussianDistribution))
end

# Infer the drift
p_m_given_M_d = estimateDrift(x_uninf, gam_0) # p(m | M_d)
println("Estimate for p(m | M_d):\n", p_m_given_M_d) # Print estimation result
```

We can now evaluate these distributions at $m=0$ and subtract the results to obtain the Bayes factor in decibels.

```julia
# Evaluate the logpfds at m=0
db_p_m_is_0_given_x_M_d = 10*log(10, pdf(Normal(mean(p_m_given_x_M_d)[1], sqrt(var(p_m_given_x_M_d)[1,1])), 0))
db_p_m_is_0_given_M_d = 10*log(10, pdf(Normal(mean(p_m_given_M_d)[1], sqrt(var(p_m_given_M_d)[1,1])), 0))

# Subtract the results
db_BF_ds = db_p_m_is_0_given_x_M_d - db_p_m_is_0_given_M_d
println("Bayes factor for preference of M_s over M_d (dB):\n", db_BF_ds)
```

A positive number indicates the number of decibels by which the simple model is preferred over the complex model. If the number is negative, then the complex model is preferred. You can play around with the actual value for the drift (`m`) and the prior over the step size (`gam_0`) to get a feel for the behaviour of the Bayes factor.

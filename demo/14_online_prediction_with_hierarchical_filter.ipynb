#Online prediction with a hierarchical filter

In this demo we use a hierarchical Gaussian filter to learn from and make online predictions for real data. This demo is inspired by the work of (Mathys, 2012). We model our data with a hierarchical filter. For more background on hierarchical filters you can take a look at the hierarchical filtering demo.

We use 180 daily measurements of the currency exchange between euros and dollars. For each day we execute a predict-measure-learn loop to estimate and learn from the next day. More specifically, we

1. predict the exchange rate for the next day based on previous samples and our hidden state estimation;
2. observe the value for the next day;
3. update the hidden state based on the newly observed sample.

##The generative model

We start again by building the generative model:

<img src="files/images/14_part1_hierarchical_filter_generative.png">

Note that this is the same model as in the previous demo, but with a shorter notation for the gain and sample chains, so we can easily extend it later on.

```julia
using PyPlot
using ForneyLab

# Wrap the generative model in a function so we can later initialize and rewire it
function buildGenerativeModel()
    # Build the generative model
    graph = FactorGraph()

    # Layer 1
    l1_x_k_min_node = TerminalNode(name="l1_x_k_min") # Observed signal
    l1_x_k_node = TerminalNode(vague(GaussianDistribution), name="l1_x_k")
    l1_add_node = AdditionNode(name="l1_add")
    l1_exp_node = ExponentialNode(name="l1_exp")
    l1_g_node = GaussianNode(form="precision", m=0.0, name="l1_g")
    Edge(l1_x_k_min_node.out, l1_add_node.in1)
    Edge(l1_add_node.out, l1_x_k_node.out)
    Edge(l1_exp_node.out, l1_g_node.precision, GammaDistribution)
    Edge(l1_g_node.out, l1_add_node.in2, GaussianDistribution)

    # Layer 2
    l2_x_k_min_node = TerminalNode(name="l2_x_k_min")
    l2_x_k_node = TerminalNode(vague(GaussianDistribution), name="l2_x_k")
    l2_add_node = AdditionNode(name="l2_add")
    l2_eq_node = EqualityNode(name="l2_eq")
    l2_exp_node = ExponentialNode(name="l2_exp")
    l2_g_node = GaussianNode(form="precision", m=0.0, name="l2_g")
    Edge(l2_eq_node.interfaces[3], l1_exp_node.in1)
    Edge(l2_add_node.out, l2_eq_node.interfaces[1])
    Edge(l2_eq_node.interfaces[2], l2_x_k_node.out)
    Edge(l2_x_k_min_node.out, l2_add_node.in1)
    Edge(l2_exp_node.out, l2_g_node.precision, GammaDistribution)
    Edge(l2_g_node.out, l2_add_node.in2, GaussianDistribution)

    # Layer 3
    theta_k_min_node = TerminalNode(name="theta_k_min")
    theta_k_node = TerminalNode(vague(GaussianDistribution), name="theta_k")
    l3_eq_node = EqualityNode(name="l3_eq")
    Edge(l3_eq_node.interfaces[3], l2_exp_node.in1)
    Edge(l3_eq_node.interfaces[2], theta_k_node.out)
    Edge(theta_k_min_node.out, l3_eq_node.interfaces[1]);
    
    return graph
end;
```

## Dataset

Next we load and visualize the dataset. These are the dayly exchance rates from dollars to euros from June 22 to December 18, 2014, obtained from `www.oanda.com`.

```julia
# Load the dataset
data = readdlm("data/dollar_euro_exchange_rate.csv", ';')
x = float(reverse(data[:,2])) # Extract values and order from past to present
N = length(x)

# Visualize the data
scatter(1:N, x)
xlabel("day")
ylabel("exchange rate")
grid(true)
title("Exchange rate from dollars to euros");
```

## Interconnections for prediction phase

We use the generative model as a basis and prepare the graph for prediction of a new sample. We set initial priors, buffers and generate a schedule that we will later execute upon inference.

```julia
# Initialize generative model
buildGenerativeModel()
prediction_graph = getCurrentGraph()

# Priors
theta_0 = GaussianDistribution(m=0.0, V=0.1)
l2_x_0 = GaussianDistribution(m=0.0, V=0.1)
l1_x_0 = GaussianDistribution(m=0.735, V=0.05) # Sample prior
findNodeByName("theta_k_min").value = theta_0
findNodeByName("l2_x_k_min").value = l2_x_0

# Buffer for obtaining samples
l1_x_k_min_buffer = setReadBuffer(findNodeByName("l1_x_k_min"), deepcopy([l1_x_0, x[1:end-1]]))

# Initialize write buffer for prediction
l1_x_k_est_buffer = setWriteBuffer(findNodeByName("l1_x_k").out.partner) # Buffer for storing prediction

# Generate a schedule for prediction
generateSchedule!(findNodeByName("l1_x_k").out.partner)

# Get expectations where required
predict_schedule = prediction_graph.active_scheme.factorization[1].internal_schedule
setSummaryOperation!(predict_schedule, findNodeByName("l2_exp").out, :sumproduct_expectation)
setSummaryOperation!(predict_schedule, findNodeByName("l1_exp").out, :sumproduct_expectation);

# graphViz(prediction_graph) # Uncomment to inspect
```

## Interconnections for measure and learn phase

We prepare a second graph object that we use for learning. It has the same structure as the prediction setup, but is now interconnected for learning the hidden states. First it observes the true value for $x^{(1)}_k$, and from that and the priors it estimates the hidden states at timepoint $k$.

Note that, as in the previous demo, we introduce a mean field factorization to accomplish the propagation towards the hidden states. Also note that we defined timewraps between the two graphs. This causes our estimation of the hidden states to become the new priors for rate prediction in the consecutive step.

```julia
# Initialize generative model; this starts a new graph.
buildGenerativeModel()
estimation_graph = getCurrentGraph()

# Initialize the same priors as before
findNodeByName("theta_k_min").value = theta_0
findNodeByName("l2_x_k_min").value = l2_x_0

# Buffers for obtaining samples
l1_x_k_min_buffer = setReadBuffer(findNodeByName("l1_x_k_min"), deepcopy([l1_x_0, x[1:end-1]]))
l1_x_k_buffer = setReadBuffer(findNodeByName("l1_x_k"), deepcopy(x))

# Initialize write buffer for state estimation
l2_x_k_est_buffer = setWriteBuffer(findNodeByName("l2_x_k").out.partner)
theta_k_est_buffer = setWriteBuffer(findNodeByName("theta_k").out.partner)

# Reset estimation graph
addTimeWrap(findNodeByName("theta_k", estimation_graph), findNodeByName("theta_k_min", estimation_graph))
addTimeWrap(findNodeByName("l2_x_k", estimation_graph), findNodeByName("l2_x_k_min", estimation_graph))
# Reset prediction graph
addTimeWrap(findNodeByName("theta_k", estimation_graph), findNodeByName("theta_k_min", prediction_graph))
addTimeWrap(findNodeByName("l2_x_k", estimation_graph), findNodeByName("l2_x_k_min", prediction_graph))

# Mean field factorization
factorizeMeanField!()
setVagueQDistributions!()

# Generate a schedule for parameter estimation
generateSchedule!();

# for subgraph in estimation_graph.active_scheme.factorization # Uncomment to inspect
#    graphViz(subgraph)
# end
```

# Perform inference and online learing

When we execute the predict-measure-learn loop by iterating between these graphs. We alternate prediction and state estimation. The buffers and timewraps we configured between the two graphs take care of all the accounting and reset the system for the next iteration.

```julia
while !isempty(l1_x_k_buffer)
    step(prediction_graph) # Predict
    step(estimation_graph, n_iterations=10) # Measure and learn
end

# Plot prediction results
x_est_m = Float64[mean(x_k_est)[1] for x_k_est in l1_x_k_est_buffer]
x_est_V = Float64[var(x_k_est)[1,1] for x_k_est in l1_x_k_est_buffer]
plot(1:N, x_est_m, color="red", label="predicted rate", alpha=0.3)
fill_between(1:N, x_est_m-sqrt(x_est_V), x_est_m+sqrt(x_est_V), color="red", alpha=0.5)
plot(1:N, x, linestyle="--", color="blue", label="actual rate")
xlabel("day")
ylabel("exchance rate")
ylim(minimum(x)-0.005, maximum(x)+0.005)
grid(true)
legend(loc="upper left")
title("Exchange rate prediction and actual value for consecutive days");
```

As can be seen from the graph above, our prediction mean lags one day behind the true values for the exchange rate. This is because we modeled the process as a zero-mean random walk. For a zero-mean random walk, the best prediction for the next day is the current value. So in effect we are only learning the volatility of the exchange rate. We can extend our model by incorporating trends. We model trends by introducing a variable mean for the lowest level Gaussian node. This mean can be determined by a model that takes previous values or knowledge about the market into account. Factor graphs can be easily extended upon as will be illustrated in the next demo, where we execute the same procedure on the same data with a model that estimates the trend as well.

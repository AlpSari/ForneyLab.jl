#Online prediction demo, part 2

As explained in part one of this demo, we will use a hierarchical Gaussian filter to learn from and make online predictions for a currency exchange rate. In part one we estimated the precision of our exchange rate. In part two we additionally model the trend with a hierarchical Gaussian filter on the mean value.

## The generative model
We start again with the generative model:

<img src="files/images/14_part2_hierarchical_filter_generative.png">

```julia
using PyPlot
using ForneyLab

# Wrap the generative model in a function so we can later initialize and rewire it
function buildGenerativeModel()
    # Build the generative model
    graph = FactorGraph()

    # Layer 1
    l1_x_k_min_node = TerminalNode(name="l1_x_k_min") # Observed signal
    l1_x_k_node = TerminalNode(vague(GaussianDistribution), name="l1_x_k")
    l1_add_node = AdditionNode(name="l1_add")
    l1_log_node = LogarithmicNode(name="l1_log")
    l1_g_node = GaussianNode(form="precision", name="l1_g")
    Edge(l1_x_k_min_node.out, l1_add_node.in1)
    Edge(l1_add_node.out, l1_x_k_node.out)
    Edge(l1_log_node.out, l1_g_node.precision, GammaDistribution)
    Edge(l1_g_node.out, l1_add_node.in2, GaussianDistribution)

    # Layer 2 for precision (volatility)
    l2_prec_x_k_min_node = TerminalNode(name="l2_prec_x_k_min")
    l2_prec_x_k_node = TerminalNode(vague(GaussianDistribution), name="l2_prec_x_k")
    l2_prec_add_node = AdditionNode(name="l2_prec_add")
    l2_prec_eq_node = EqualityNode(name="l2_prec_eq")
    l2_prec_log_node = LogarithmicNode(name="l2_prec_log")
    l2_prec_g_node = GaussianNode(form="precision", m=0.0, name="l2_prec_g")
    Edge(l2_prec_eq_node.interfaces[3], l1_log_node.in1)
    Edge(l2_prec_add_node.out, l2_prec_eq_node.interfaces[1])
    Edge(l2_prec_eq_node.interfaces[2], l2_prec_x_k_node.out)
    Edge(l2_prec_x_k_min_node.out, l2_prec_add_node.in1)
    Edge(l2_prec_log_node.out, l2_prec_g_node.precision, GammaDistribution)
    Edge(l2_prec_g_node.out, l2_prec_add_node.in2, GaussianDistribution)

    # Layer 2 for mean (trend)
    l2_mean_x_k_min_node = TerminalNode(name="l2_mean_x_k_min")
    l2_mean_x_k_node = TerminalNode(vague(GaussianDistribution), name="l2_mean_x_k")
    l2_mean_add_node = AdditionNode(name="l2_mean_add")
    l2_mean_eq_node = EqualityNode(name="l2_mean_eq")
    l2_mean_log_node = LogarithmicNode(name="l2_mean_log")
    l2_mean_g_node = GaussianNode(form="precision", m=0.0, name="l2_mean_g")
    Edge(l2_mean_eq_node.interfaces[3], l1_g_node.mean, GaussianDistribution)
    Edge(l2_mean_add_node.out, l2_mean_eq_node.interfaces[1])
    Edge(l2_mean_eq_node.interfaces[2], l2_mean_x_k_node.out)
    Edge(l2_mean_x_k_min_node.out, l2_mean_add_node.in1)
    Edge(l2_mean_log_node.out, l2_mean_g_node.precision, GammaDistribution)
    Edge(l2_mean_g_node.out, l2_mean_add_node.in2, GaussianDistribution)

    # Layer 3 for precision
    theta_prec_k_min_node = TerminalNode(name="theta_prec_k_min")
    theta_prec_k_node = TerminalNode(vague(GaussianDistribution), name="theta_prec_k")
    l3_prec_eq_node = EqualityNode(name="l3_prec_eq")
    Edge(l3_prec_eq_node.interfaces[3], l2_prec_log_node.in1)
    Edge(l3_prec_eq_node.interfaces[2], theta_prec_k_node.out)
    Edge(theta_prec_k_min_node.out, l3_prec_eq_node.interfaces[1]);
    
    # Layer 3 for mean
    theta_mean_k_min_node = TerminalNode(name="theta_mean_k_min")
    theta_mean_k_node = TerminalNode(vague(GaussianDistribution), name="theta_mean_k")
    l3_mean_eq_node = EqualityNode(name="l3_mean_eq")
    Edge(l3_mean_eq_node.interfaces[3], l2_mean_log_node.in1)
    Edge(l3_mean_eq_node.interfaces[2], theta_mean_k_node.out)
    Edge(theta_mean_k_min_node.out, l3_mean_eq_node.interfaces[1]);

    return graph
end;

# Load the dataset
data = readdlm("data/dollar_euro_exchange_rate.csv", ';')
x = float(reverse(data[:,2])) # Extract values and order from past to present
N = length(x)
```

## Interconnections for prediction phase

We use the generative model as a basis and prepare the graph for prediction of a new sample. We set initial priors, buffers and generate a schedule that we will later execute upon inference.

```julia
# Initialize generative model
buildGenerativeModel()
prediction_graph = getCurrentGraph()

# Priors
theta_prec_0 = GaussianDistribution(m=0.0, V=0.1)
theta_mean_0 = GaussianDistribution(m=0.0, V=0.1)
l2_prec_x_0 = GaussianDistribution(m=0.0, V=0.1)
l2_mean_x_0 = GaussianDistribution(m=0.0, V=0.1)
l1_x_0 = GaussianDistribution(m=0.735, V=0.05) # Sample prior
findNodeByName("theta_prec_k_min").value = theta_prec_0
findNodeByName("theta_mean_k_min").value = theta_mean_0
findNodeByName("l2_prec_x_k_min").value = l2_prec_x_0
findNodeByName("l2_mean_x_k_min").value = l2_mean_x_0

# Buffer for obtaining samples
l1_x_k_min_buffer = setReadBuffer(findNodeByName("l1_x_k_min"), deepcopy([l1_x_0, x[1:end-1]]))

# Initialize write buffer for prediction
l1_x_k_est_buffer = setWriteBuffer(findNodeByName("l1_x_k").out.partner) # Buffer for storing prediction

# Generate a schedule for prediction
generateSchedule!(findNodeByName("l1_x_k").out.partner)

# Get expectations where required
predict_schedule = prediction_graph.factorization[1].internal_schedule
setSummaryOperation!(predict_schedule, findNodeByName("l2_prec_log").out, :sumproduct_expectation)
setSummaryOperation!(predict_schedule, findNodeByName("l2_mean_log").out, :sumproduct_expectation)
setSummaryOperation!(predict_schedule, findNodeByName("l1_g").precision.partner, :sumproduct_expectation);
setSummaryOperation!(predict_schedule, findNodeByName("l1_g").mean.partner, :sumproduct_expectation);

# graphViz(prediction_graph) # Uncomment to inspect
```

## Interconnections for measure and learn phase

We prepare a second graph object that we use for learning. It has the same structure as the prediction setup, but is now interconnected for learning the hidden states. First it observes the true value for $x^{(1)}_k$, and from that and the priors it estimates the hidden states at timepoint $k$. Note that we added timewarps for the trend estimation.

```julia
# Initialize generative model; this starts a new graph.
buildGenerativeModel()
estimation_graph = getCurrentGraph()

# Initialize the same priors as before
findNodeByName("theta_prec_k_min").value = theta_prec_0
findNodeByName("theta_mean_k_min").value = theta_mean_0
findNodeByName("l2_prec_x_k_min").value = l2_prec_x_0
findNodeByName("l2_mean_x_k_min").value = l2_mean_x_0

# Buffers for obtaining samples
l1_x_k_min_buffer = setReadBuffer(findNodeByName("l1_x_k_min"), deepcopy([l1_x_0, x[1:end-1]]))
l1_x_k_buffer = setReadBuffer(findNodeByName("l1_x_k"), deepcopy(x))

# Initialize write buffer for state estimation
l2_prec_x_k_est_buffer = setWriteBuffer(findNodeByName("l2_prec_x_k").out.partner)
l2_mean_x_k_est_buffer = setWriteBuffer(findNodeByName("l2_mean_x_k").out.partner)
theta_prec_k_est_buffer = setWriteBuffer(findNodeByName("theta_prec_k").out.partner)
theta_mean_k_est_buffer = setWriteBuffer(findNodeByName("theta_mean_k").out.partner)

# Reset estimation graph
addTimeWrap(findNodeByName("theta_prec_k", estimation_graph), findNodeByName("theta_prec_k_min", estimation_graph))
addTimeWrap(findNodeByName("theta_mean_k", estimation_graph), findNodeByName("theta_mean_k_min", estimation_graph))
addTimeWrap(findNodeByName("l2_prec_x_k", estimation_graph), findNodeByName("l2_prec_x_k_min", estimation_graph))
addTimeWrap(findNodeByName("l2_mean_x_k", estimation_graph), findNodeByName("l2_mean_x_k_min", estimation_graph))
# Reset prediction graph
addTimeWrap(findNodeByName("theta_prec_k", estimation_graph), findNodeByName("theta_prec_k_min", prediction_graph))
addTimeWrap(findNodeByName("theta_mean_k", estimation_graph), findNodeByName("theta_mean_k_min", prediction_graph))
addTimeWrap(findNodeByName("l2_prec_x_k", estimation_graph), findNodeByName("l2_prec_x_k_min", prediction_graph))
addTimeWrap(findNodeByName("l2_mean_x_k", estimation_graph), findNodeByName("l2_mean_x_k_min", prediction_graph))

# Mean field factorization
factorizeMeanField!()
setVagueMarginals!()

# Generate a schedule for state estimation
generateSchedule!();

# for subgraph in estimation_graph.factorization # Uncomment to inspect
#    graphViz(subgraph)
# end
```

# Perform inference and online learing

When we execute the predict-measure-learn loop by iterating between these graphs. We alternate prediction and state estimation. The buffers and timewraps we configured between the two graphs take care of all the accounting and reset the system for the next iteration.

```julia
while !isempty(l1_x_k_buffer)
    step(prediction_graph) # Predict
    step(estimation_graph, n_iterations=10) # Measure and learn
end

# Plot prediction results
x_est_m = Float64[mean(x_k_est)[1] for x_k_est in l1_x_k_est_buffer]
x_est_V = Float64[var(x_k_est)[1,1] for x_k_est in l1_x_k_est_buffer]
plot(1:N, x_est_m, color="red", label="predicted rate")
fill_between(1:N, x_est_m-sqrt(x_est_V), x_est_m+sqrt(x_est_V), color="red", alpha=0.2)
plot(1:N, x, linestyle="--", color="blue", label="actual rate")
xlabel("day")
ylabel("exchance rate")
ylim(minimum(x)-0.025, maximum(x)+0.025)
grid(true)
legend(loc="upper left")
title("Exchange rate prediction and actual value for consecutive days");
```

From this graph it becomes clear that the uncertainty bands are way bigger than before, meaning the model makes more conservative estimates. Also the estimates no longer structurally lag behind the data. Do note however that this model overshoots the actual data at peak values. Once a trend is detected, the model assumes the trend will continue.
